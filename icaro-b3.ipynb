```python
# Import required libraries
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock tickers for the two portfolios
portfolio1_tickers = ['HGLG11.SA', 'BPAC11.SA', 'ITUB4.SA']
portfolio2_tickers = ['IVVB11.SA']
# portfolio3_tickers = ['META']

# Download the stock data for the last 2 years
tickers = portfolio1_tickers + portfolio2_tickers 
data = yf.download(tickers, start='2021-01-01', end='2024-12-01', auto_adjust=True).get("Close")

# Normalize the data to show growth
normalized_data = data / data.iloc[0]

# Calculate the average performance of each portfolio
portfolio1 = normalized_data[portfolio1_tickers].mean(axis=1)
portfolio2 = normalized_data[portfolio2_tickers].mean(axis=1)
# portfolio3 = normalized_data[portfolio3_tickers].mean(axis=1)

# Plot the normalized data
plt.figure(figsize=(14, 7))
plt.plot(portfolio1, label='Portfolio 1: HGLG11 + BPAC11 + ITUB4')
plt.plot(portfolio2, label='Portfolio 2: IVVB11')
# plt.plot(portfolio3, label='Portfolio 3: META')

# Add title and labels
plt.title('Portfolio Performance Comparison')
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.legend()
plt.grid(True)
plt.show()
```
