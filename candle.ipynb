```python
# Import required libraries
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock tickers
high_growth_tickers = ["AMZN", "TSLA"]
stable_valuation_tickers = ["KO", "XOM"]

# Download the stock data for the last 2 years
tickers = high_growth_tickers + stable_valuation_tickers
data = yf.download(tickers, start="2021-01-01", end="2024-12-01")["Close"]

# Normalize the data to show growth
normalized_data = data / data.iloc[0]

# Plot the normalized data
plt.figure(figsize=(14, 7))
for ticker in high_growth_tickers:
    plt.plot(normalized_data[ticker], label=f"High Growth: {ticker}")
for ticker in stable_valuation_tickers:
    plt.plot(normalized_data[ticker], label=f"Stable Valuation: {ticker}")

# Add title and labels
plt.title("High Growth vs Stable Valuation Companies (Last 2 Years)")
plt.xlabel("Date")
plt.ylabel("Normalized Price")
plt.legend()
plt.grid(True)
plt.show()
```
